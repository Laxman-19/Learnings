```python
original_title = []
description = []
genre = []
movie_id = []
original_language = []
release_date = []
english_title = []


for page in range(1,5):
    res = requests.get(f'https://api.themoviedb.org/3/movie/top_rated?api_key=8265bd1679663a7ea12ac168da84d2e8&language=en-US&page={page}')
    response1 = json.loads(res.text)

    
    
    list = response1['results']
    
    for each_movie in list:
        original_title.append(each_movie['original_title'])
        description.append(each_movie['overview'])
        genre.append(each_movie['genre_ids'])
        movie_id.append(each_movie['id'])
        original_language.append(each_movie['original_language'])
        release_date.append(each_movie['release_date'])
        english_title.append(each_movie['title'])
 
    print('----------------------------------------------------------------')
```

```python
movie_data = {}

movie_data = {
    'Movie ID' : movie_id,
    'Genre ID' : genre,
    'Original Title' : original_title,
    'English Title' : english_title,
    'Language Code' : original_language,
    'Release Date' : release_date,
    'Description' : description
}

movie_data
```

```python
movie_df = pd.DataFrame(movie_data)
```

```python
res = requests.get(f'https://api.themoviedb.org/3/genre/movie/list?api_key=8265bd1679663a7ea12ac168da84d2e8&language=en-US')
response2 = json.loads(res.text)

response2
```

```python
dict = response2['genres']
```

```python
new_dict = {}

for i in dict:
    new_dict[(i['id'])] = i['name']
    
new_dict
```

```python
def replace_genre(list):
    result = ''
    aaa = []
    
    for i in list:
        if i in new_dict:
            aaa.append(new_dict[i])
            
    result = ' '.join(aaa)
    return result

a = [18, 80]
replace_genre(a)
```

```python
movie_df['Genre Name'] = movie_df['Genre ID'].apply(replace_genre)
```
