# Adding Target Variable
- You can do it in 2 ways.
- Both the ways are shown below.

### First way:

```python
binned = pd.cut(new_data['avg'], bins = [0, 60, 100, 115], labels = ['Normal', 'Osteopenia', 'Osteoporotic'])

new_data['OUTPUT'] = binned
```

```python
new_data.head()
```

### Second way:

```python
def tv(i):
    if (i<60):
        return 'Normal'
    elif (i>=60 and i<=100):
        return 'Osteopenia'
    else:
        return 'Osteoporotic'
```

```python
new_data['OUTPUT_2'] = new_data['avg'].astype(float).apply(tv)
```

```python
new_data.head()
```

```python
new_data['OUTPUT'].value_counts()
```

```python
new_data['OUTPUT_2'].value_counts()
```

# Mapping

```python
mapped_df = new_data.copy()
```

```python
new_data.head()
```

```python
mapped_df.head()
```

```python
mapped_df.SEX = mapped_df.SEX.map( {'Male' : 1, 'Female' : 0} )

mapped_df['ASSO MEDICAL PROB'] = mapped_df['ASSO MEDICAL PROB'].map( {'No' : 0,
                                                                      'Yes(Diabetes)' : 1,
                                                                      'Yes(Diabetes, BP)' : 2,
                                                                      'Yes(BP)' : 3,
                                                                      'Yes(Diabetes, Heart Blockage)' : 4,
                                                                      'Kidney Stone' : 5,
                                                                      'Yes(Increased Heart Rate)' : 6,
                                                                      'Yes(Diabetes, Kidney Stone)' : 7, } )

mapped_df['H/O INJURY/SURGERY'] = mapped_df['H/O INJURY/SURGERY'].map( {'No' : 0,
                                                                        'Uterus Surgery' : 1,
                                                                        'Vericose Vein Surgery' : 2,
                                                                        'Kidney Stone Opreration' : 3,
                                                                        'Yes(Diverticulities)' : 4,
                                                                        'Shouler Surgery' : 5,
                                                                        'Knee Surgery' : 6,
                                                                        'Yes(Open Heart Surgery)' : 7, } )

mapped_df['DRUG HISTORY'] = mapped_df['DRUG HISTORY'].map( {'No' : 0,
                                                            'Yes' : 1,
                                                            'Yes(Ecosprin)' : 2, } )
```

```python
mapped_df.head()
```

# Training

```python
x = mapped_df.drop(['avg', 'OUTPUT', 'OUTPUT_2'], axis = 1)
y = mapped_df[['OUTPUT']]
```

```python
x.head()
```

```python
y.head()
```

```python
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier()
model.fit(x, y)

model.predict( [ [29, 1, 0, 0, 0] ] )
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2)
```

```python
from sklearn.preprocessing import StandardScaler

sc_x = StandardScaler()
x_train = sc_x.fit_transform(x_train)
x_test = sc_x.transform(x_test)
```

# Creating Model

```python
from sklearn.tree import DecisionTreeClassifier

model2 = DecisionTreeClassifier()
model2.fit(x_train, y_train)

predictions = model2.predict(x_test)

score = accuracy_score(y_test, predictions)
score
```

```python
from sklearn.linear_model import LogisticRegression

model3 = LogisticRegression()
model3.fit(x_train, y_train)

predictions = model3.predict(x_test)

score = accuracy_score(y_test, predictions)
score
```
